```
# Importaciones necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```
# Creamos la funcion de coste

def func(array):
  z = np.sin((1/2)*(array[0])**2 - (1/4)*(array[1])**2 + 3) * np.cos(2*array[0] + 1 - (np.e)**array[1])
  return z
```

```
# Crear vectores aleatorios de una longitud considerable de puntos, 
# será como la escala del mapa.

resolution = 100   # Será la extensión del arreglo que vamos a crear

_X = np.linspace(-2, 2, resolution)
_Y = np.linspace(-2, 2, resolution)


# # Popularemos _Y con numeros aleatorios:
# for i in range(100):
#   _Y[i] = np.random.rand()


# Ahora hay que popular el vector _Z
_Z = np.zeros((resolution, resolution))  # Primero necesitamos crearla

for indexX, x in enumerate(_X):
  for indexY, y in enumerate(_Y):
    aux_array = [x, y]
    _Z[indexY][indexX] = func(aux_array)


plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.show()
```

```
# Creamos un punto en una parte aleatoria dele plano 
Theta = np.random.rand(2) * 4 - 2

plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.plot(Theta[0], Theta[1], "o", c="red")
plt.show()
```

```
# Calculamas las derivadas por el metodo de diferencias mínimas

_T = np.copy(Theta)

h = 0.001

grad = np.zeros(2)
lr = 0.001
```

```

# Para cada elemento dedl array calcula su derivada
for indexTheta, theta in enumerate(Theta):

  _T[indexTheta] = _T[indexTheta] + h
  print(theta)
  
  deriv = (func(_T) - func(Theta)) / h

  grad[indexTheta] = deriv
  print(grad)
Theta = Theta - lr * grad


plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.plot(grad[0], grad[1], "o", c="white")
plt.plot(Theta[0], Theta[1], "o", c="red")
plt.show()
```
